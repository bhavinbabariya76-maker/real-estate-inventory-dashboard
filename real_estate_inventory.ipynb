```python
import pandas as pd
```

```python
df = pd.read_csv("real_estate_inventory.csv")
```

```python
df.head()
```

```python
df["Inventory_Left"] = (df["total_units"] - df["units_sold"])
```

```python
df["Absorption_Rate"] = (df["units_sold"] /df["total_units"]) * 100
```

```python
df.sort_values("units_sold",ascending=False)
```

```python
df.sort_values("Inventory_Left",ascending=False)
```

```python
df.groupby("city")["Inventory_Left"].sum()
```

```python
df[df["Absorption_Rate"] < 50]
```

```python
import matplotlib.pyplot as plt
```

```python
df.plot(
    x="city",
    y="Inventory_Left",
    kind="bar"
)

plt.title("Inventory Left by Project")
plt.show()
```

```python
df.plot(
    x="city",
    y="Absorption_Rate",
    kind="bar"
)

plt.title("Absorption Rate")
plt.show()
```

```python
df.groupby("city")[
    "units_sold"
].sum().plot(kind="bar")

plt.title("Units Sold by City")
plt.show()
```
